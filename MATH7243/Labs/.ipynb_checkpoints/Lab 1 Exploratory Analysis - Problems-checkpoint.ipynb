# Exploratory Analysis

## Problems:

Load the NYC AirBnB Truncated Dataset. This dataset is a mirror of the full NYC AirBnB dataset found at Kaggle, but only contains the first 10,000 entries. 

https://www.kaggle.com/dgomonov/new-york-city-airbnb-open-data

For the numerical features, 

1) Display histograms for the numerical features. 
2) Construct the scatter plots of price with each of the numerical features.
3) Display the correlation histogram.
4) Which numerical feature appears to be most predictive of the price?

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv("https://raw.githubusercontent.com/tipthederiver/Math-7243-2020/master/Datasets/NYCAirBnB/train.csv")
data
```

```python
display(type(data))
display(data.head(5))
display(data.tail(5))
print("Our datasets has dimensions", data.shape)
```

```python
names = list(data)
print(names)
```

```python
data.info()
```

# From the above information Numerical Features can be distinguished against Non-numerical features based on their datatypes. int64 and float 64 are numerical features while object is non numerical feature.

Numerical Features : latitude, longitude, price, minimum_nights, number_of_reviews, reviews_per_month, calculated_host_listings_count, availability_365

Non-Numerical Features: name, host_id, host_name, neighbourhood_group, neighbourhood, room_type, last_review  

# Histograms for all the numerical features are as follows:

```python
from matplotlib import pyplot as plt
fig, ax = plt.subplots()

ax.hist(data.latitude,bins=50, rwidth=0.8)
ax.set_xlabel("Latitude", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("In which latitude AirBNB rooms can be found?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(data.longitude,bins=50, rwidth=0.8)
ax.set_xlabel("Longitude", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("In which longitude AirBNB rooms can be found?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

p=data.price<5000
ax.hist(data.price[p],bins=50, rwidth=0.8)
ax.set_xlabel("Price", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("How much do AirBNB rooms cost?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(data.minimum_nights[p],bins=50, rwidth=0.8)
ax.set_xlabel("Minimum Nights", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("How long is the minimum stay?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(data.number_of_reviews[p],bins=50, rwidth=0.8)
ax.set_xlabel("Number of Reviews", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("How many reviews does the AirBNB rooms get?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(data.reviews_per_month[p],bins=50, rwidth=0.8)
ax.set_xlabel("Reviews Per Month", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("How many reviews does AirBNB rooms get in a month?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(data.calculated_host_listings_count[p],bins=50, rwidth=0.8)
ax.set_xlabel("Host Listings Count", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("How many hosts are listed?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(data.availability_365,bins=10, rwidth=0.5)
ax.set_xlabel("Number of Reviews", fontsize=15)
ax.set_ylabel("Count", fontsize=15)
ax.set_title("What's the availability of AirBNB's rooms in an year?", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
nums = data.select_dtypes(include=['int64','float64'])
display(nums.head(5))
numsc = nums[p]
display(numsc.head(5))
```

```python
nnames = list(numsc)
len(nnames)
```

```python
f, ax = plt.subplots(3, 3, sharey = True)
f.set_size_inches(15,10)
f.tight_layout()

ax = ax.reshape(9)

for i in range(0,9):
    ax[i].hist(numsc[nnames[i]],numsc.price[i], rwidth = 0.9)
    ax[i].set_title(nnames[i], fontsize=12)
```

# Scatter Plots of Price against all numerical features are as follows:

```python
fig, ax = plt.subplots()

ax.plot(numsc.latitude[p],numsc.price[p],'o')
ax.set_xlabel("Latitude", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Latitude vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(numsc.longitude[p],numsc.price[p],'o')
ax.set_xlabel("Longitude", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Longitude vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(numsc.minimum_nights[p],numsc.price[p],'o')
ax.set_xlabel("Minimum Nights", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Minimum Nights vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(numsc.number_of_reviews[p],numsc.price[p],'o')
ax.set_xlabel("number_of_reviews", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Number of Reviews vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(numsc.reviews_per_month[p],numsc.price[p],'o')
ax.set_xlabel("Reviews Per Month", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Reviews Per Month vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(numsc.calculated_host_listings_count[p],numsc.price[p],'o')
ax.set_xlabel("Host Listings Count", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Host Listings Count vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(numsc.availability_365[p],numsc.price[p],'o')
ax.set_xlabel("Availability", fontsize=15)
ax.set_ylabel("Price", fontsize=15)
ax.set_title("Availability vs Price", fontsize=20)
fig.set_size_inches(8, 5)
plt.show()
```

# Scatterplots of all features with price as subplots for better comparison

```python
f, axes = plt.subplots(3, 3, sharey = True)
f.set_size_inches(15,10)
f.tight_layout()

axes = axes.reshape(9)

for i in range(0,9):
    axes[i].plot(numsc[nnames[i]],numsc.price,'o')
    axes[i].set_title(nnames[i], fontsize=12)
```

# Correlation Plot

```python
import seaborn as sns
fig,ax = plt.subplots(figsize=(14, 10))

sns.heatmap(numsc.corr(), ax=ax, linewidths=0.05,cmap="magma")
plt.show()
```

```python
sns.pairplot(numsc)
plt.show()
```

```python
numsc.corr()
```

According to the above correlation heatmap and correlation matrix, none of the numerical features appears to be most predictive of the price as all the values are below 0.1. In comprison to other numerical features, latitude is the most predictive of price followed by availability.
