```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g_x(x, beta):
    return beta[0] + np.sin(beta[1] * x) + np.cos(beta[2] * x)
```

```python
def RSS_derivative_w_beta(y, x, beta):
    rss_b = np.zeros(3)
    vector_one = np.ones_like(x)
    y_minus_gx = y - g_x(x, beta)
    rss_b[0] = -np.sum(2 * y_minus_gx * vector_one)
    rss_b[1] = -np.sum(2 * y_minus_gx * np.cos(beta[1] * x) * x)
    rss_b[2] = np.sum(2 * y_minus_gx * np.sin(beta[2] * x) * x)
    return rss_b
```

```python
x = np.arange(0, 19, 2)
y = np.array([2.85, 1.5, 0.49, 1.57, 1.9, 0.6, 0.38, 2.33, 1.65, 0.3])
```

```python
eta = 0.01
epochs = 50
beta = [0.1, 2.5, 0.5]
rss_b = RSS_derivative_w_beta(y, x, beta)
print(rss_b)
plt.plot(x, y,label = "Original Data")
for epoch in range(1, epochs):
    rss_b = RSS_derivative_w_beta(y, x, beta)
    beta = beta - (eta / np.sqrt(epoch)) * rss_b
    if epoch % 8 == 0:
        plt.plot(x, g_x(x, beta), label=f"epoch:{epoch}")
plt.legend()
plt.title("GD by Minimizing RSS Loss")
plt.show()
```
