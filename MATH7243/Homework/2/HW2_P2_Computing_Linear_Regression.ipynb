```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1.2, 3.2, 5.1, 3.5, 2.6])
y = np.array([7.8, 1.2, 6.4, 2.6, 8.1])
```

```python
X = np.ones_like(x)
X = np.concatenate([X[:, np.newaxis], x[:, np.newaxis]], axis=1)
```

```python
def parameters_for_ridge(l, x, y):
    y = y[:, np.newaxis]
    xtx = np.matmul(X.T, X)
    theta = np.matmul(np.matmul(np.linalg.inv(xtx + l ** 2), X.T), y)
    return theta
```

```python
plt.scatter(x, y, label="Original Points")
for l in [0, 1, 10]:
    theta = parameters_for_ridge(l=l, x=x, y=y)
    y_hat = np.matmul(X, theta)
    plt.plot(x, y_hat, label=f"fit with lambda = {l}")
    plt.legend()
plt.title("Linear Regression using Ridge Loss")
plt.show()
```

For RSS Loss, Lambda is Zero, which is the best fit for this function.
For Ridge Loss, when lamba is near zero, the fit line starts drifting from the best fit, when lambda is too high, slope of line changes drastically.
