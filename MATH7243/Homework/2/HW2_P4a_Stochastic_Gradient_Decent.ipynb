```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def parameters_for_ridge(x, y):
    xtx = np.matmul(X.T, X)
    theta = np.matmul(np.matmul(np.linalg.inv(xtx), X.T), y)
    return theta
```

```python
x = np.arange(0, 19, 2)
y = np.array([2.85, 1.5, 0.49, 1.57, 1.9, 0.6, 0.38, 2.33, 1.65, 0.3])
x = x[:, np.newaxis]
```

```python
ones_x = np.ones_like(x)
sin_x = np.sin(x)
cos_x = np.cos(x)
X = np.concatenate([ones_x, sin_x, cos_x], axis=1)
y = y[:, np.newaxis]
```

```python
theta = parameters_for_ridge(x=x, y=y)
y_hat = np.matmul(X, theta)
plt.plot(x, y_hat, label=f"fit")
plt.legend()
plt.scatter(x, y)
plt.title("Least Square Method")
plt.show()
```

Yes, we can solve the parameters using the least square methods.
Use X = [[ 1 sin(x^1) cos(x^1)],[ 1 sin(x^2) cos(x^2)],[ 1 sin(x^3) cos(x^3)],....,[ 1 sin(x^i) cos(x^i)]] nx3 matrix
theta = [[beta0] [beta1] [beta2]] 3x1 matrix
then Y = X.theta

Using Least Square Method
theta = (X^T.X)^-1.X^T.Y
