### Numpy

`import numpy as np`

__Numpy Array `a`:__

* __Create array from list of lists: `y = np.array(  [[1,2],[3,4]]  )
* `a.reshape(m,n,..)` - Change dimensions of array to $[m,n,...]$ , keeping data constant.
* `a.reshape(m,-1)` - Change dimensions of array so first dimension is $m$, let numpy work out last dimension.

__Numpy Matrix `A`:__

* __Create matrix from list of lists__: `y = np.matrix(  [[1,2],[3,4]]  )`
* __Transpose__: `A.T`
* __Inverse__: `A.I`
* __Matrix of 0's__: To produce a `n` by `m` matrix of `0`'s, `np.zeros([n,m])`.
* __Matrix of 1's__: To produce a `n` by `m` matrix of `0`'s, `np.ones([n,m])`.
* __Identity Matrix__: For an `n` by `m` identity matrix, `np.eye(n)`.
* __Random Matrix__: A uniform random matrix of size `n` by `m`, can be created with `np.random.random([n,m])`.


### Pandas Dataframes: For a Dataframe df

`import pandas as pd`

* __Create Dataframe from Dictionary:__ 
`d = {'Column 1': [1,2,3,4], 'Column 2': [5,6,7,8]}`
`df = pd.DataFrame(d)`

__Working with DataFrames `df`__
* `df.head(n)` Display the first `n` columns of the dataframe.
* `df.tail(n)` - Display the last `n` columns of the dataframe.
* `df.shape` - Contains the dimensions of df
* `dispaly(df)` - Shows dataframe contents formatted.
* `list(df)` - Lists column names.
* `df.info()` - For each column, number of nonempty entries and datatype.
* `df.hist()` - Display a histogram for all columns of df.
* `df.corr()` - Produce correlation matrix


### Matplotlib graphing:

`from matplotlib import pyplot as plt`

* `plt.hist(X,bins=bins, rwidth=)` - Display a histogram of `X` with `bins` columns of width `rwidth`.
* `plt.plot(X,Y)` - Display 2D line plot of X vs Y. 


#### Subplots: 
Note, in most calls we can set the fontsize by including `fontsize=`.
* `fig, ax = plt.subplots([m,n])` - Display an $m$ by $n$ grid of subplots.  Returns array of axes `ax[i]` of dim `[m,n]`.
* `fig.set_size_inches(10, 7)` - Set the dimensions of the image.
* `ax.xlabel("X Label")` - Set the x axis label for the picture.
* `ax.ylabel("Y Label")` - Set the y axis label for the picture.
* `ax.set_title("Title")` - Set the title for the picture.
